```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import hvplot.pandas
from bokeh.sampledata.autompg import autompg_clean
```

```python
# Set the column width
pd.options.display.max_colwidth = 200
# Read in the dataset. 
landings_df = pd.read_csv("Resources/Meteorite_Data_for_Charts.csv")
landings_df
```

```python
# Configure the map plot

Meteor_map_plot = landings_df.hvplot.scatter(x='reclat', y='reclong', s='year', scale= 1, by='name', 
                  legend='top', height=600, width=900)
# Display the map
Meteor_map_plot
```

```python
# Build the scatter plots for landing latitude vs. year
plt.scatter(landings_df["reclat"], landings_df["year"]) #landings_df["mass"])

# Incorporate the other graph properties
plt.title("Meteor Landing Latitude vs. Year")
plt.xlabel("Latitiude")
plt.ylabel("Year")

# Save the figure
plt.savefig("C:\Class\Challenges\Meteorite-Landings-Challenge\Resources")

# Show plot
plt.show()
```

```python
# Build the scatter plots for Landing Longitude vs. Year
plt.scatter(landings_df["reclong"], landings_df["year"]) #landings_df["mass"])

# Incorporate the other graph properties
plt.title("Meteor Landing Longitude vs. Year")
plt.xlabel("Longitude")
plt.ylabel("Year")

# Save the figure
plt.savefig("C:\Class\Challenges\Meteorite-Landings-Challenge\Resources")

# Show plot
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
grouped_metadata = landings_df["name"].value_counts()
Panda_Bar_Plot = grouped_metadata.plot(kind="bar", figsize= (10,4))
plt.title("Meteorite Landings")
plt.ylabel("Year Meteor Made landing")
plt.show
```
